```python
f = open("data/pascoa.txt")
for line in f:
    ano = int(line[4:])
    diaf = int(line[0:2])
    mesf = int(line[2:4])
    a = ano % 19
    b = ano % 4
    c = ano % 7
    d = (19 * a + 24) % 30
    e = (2 * b + 4 * c + 6 * d + 5) % 7
    pascoa = 22 + d + e
    if ano == 1954 or ano == 1981 or ano == 2049 or ano == 2076:
        pascoa -= 7
    assert 22 <= pascoa <= 56
    if pascoa <= 31:
        dia, mes = pascoa, 3
    else:
        dia, mes = pascoa - 31, 4
    print(format(diaf, "02d"), format(mesf, "02d"), format(ano, "4d"), end=' ')
    if dia != diaf or mes != mesf:
        print('!=', format(dia, "02d"), format(mes, "02d"), format(ano, "4d"))
    else:
        print(' ok')
```
