```python
from time import time
from math import sqrt

epsilon = 1e-8

testes = [float(1234567890**2)]
for p in range(-4, 6):
    testes += [10 ** p]
print(f'{"x":17} {"r":17} {"resp":17} {"dif":17} {"n":3} {"tempo":10}')
print(17*'-', 17*'-', 17*'-', 17*'-', 3*'-', 10*'-')
for x in testes:
    num_ciclos = 0

    start = time()
    raiz = x / 2
    while abs(raiz ** 2 - x) > epsilon:
        raiz -= (((raiz * raiz) - x) / (raiz + raiz))
        num_ciclos += 1
    end = time()

    resp = sqrt(x)

    print(f'{x:17.10e} {raiz:17.10e} {resp:17.10e} {raiz - resp:17.10e} {num_ciclos:3} {end - start:10.3e}')
```

```python
from random import choices

n = 4
perm = sorted(choices(range(1, 6), k=n))
print(perm)

tem_mais = True
while tem_mais:
    # Find non-increasing suffix
    i = - 1
    while i > -n and perm[i - 1] >= perm[i]:
        i -= 1
    if i <= -n:
        tem_mais = False
    else:
        # Find successor to pivot
        j = - 1
        while perm[j] <= perm[i - 1]:
            j -= 1
        perm[i - 1], perm[j] = perm[j], perm[i - 1]

        # Reverse suffix
        perm[i : ] = perm[- 1 : i - 1 : -1]
        print(perm)
```
